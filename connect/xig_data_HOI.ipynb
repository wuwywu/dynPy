## 获取数据名称
### xgi.load_xgi_data()

| 参数 | 类型 | 说明 |
| ---- | ---- | ---- |
| `dataset` | str, optional | 数据集名称；传`None`会打印所有可用数据集，名称不存在抛出`XGIError` |
| `cache` | bool, optional | 是否将数据缓存至本地，默认`True`；`False`每次重新拉取数据不缓存 |
| `read` | bool, optional | 数据源优先级，默认`False`优先联网；`True`优先读取本地文件，本地无数据再联网 |
| `path` | str, optional | 本地数据集存放目录，空字符串使用默认缓存路径，离线场景配合`read=True`使用 |
| `nodetype` | type, optional | 节点ID强制转换类型，可选`int`/`str`，默认不转换 |
| `edgetype` | type, optional | 超边ID强制转换类型，可选`int`/`str`，默认不转换 |
| `max_order` | int, optional | 超边阶数过滤，仅保留阶数≤该数值的超边；`None`保留全部超边 |

## 载入数据

```python
Hg = xgi.load_xgi_data("sfhh-conference")
```

## 查看网络性质

```python
print("节点数:", Hg.num_nodes)
print("超边数:", Hg.num_edges)
Hg.edges.members(e="0")
```

### 查看边的数据

```python
max_d = xgi.max_edge_order(Hg)
print("最大超边阶数：", max_d) # d = num_nodes-1
# 不同阶超边的数量
for d in range(0, max_d + 1):
    num = xgi.num_edges_order(Hg, d=d)
    if num > 0:
        print(f"阶数{d}：{num}条超边")
```
